```python
dfa=df.copy()
for i in df.columns[0:]:
  dfa[i].fillna(0,inplace=True)
dfa
```

```python
dfa.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
for i in dfa.columns[132:]:
  plt.hist(df[i], bins=20, edgecolor='black')  # Adjust bins as needed
  plt.title(i)
  plt.xlabel(i)
  plt.ylabel('Frequency')
  plt.grid(True)
  plt.show()

```

redshift- most of the redshift are in the range 0.5 to 1. the redshift have a continuous disyribution from 0 to 3

age- most of the sources have similar age

dust- dust content is very low

mass- most of the sources have mass between 0.75 to 10

star formation- star formation is also low.
